````python
import streamlit as st
import os, json, shutil, pickle, datetime, subprocess, random, requests, sys, time
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from googleapiclient.errors import HttpError
from google.auth.transport.requests import Request
from PIL import Image
from concurrent.futures import ThreadPoolExecutor, as_completed

# =========================
# üîë KONFIGURASI GLOBAL
# =========================
GEMINI_API_KEY = "AIzaSyAZ7SsQT63mFCs5KZH6pNRmZyixlnxYyco"
PEXELS_API_KEY = "yMF2kUa65DyusYQjXb1zZzJ2oMnaM1oKZYA5TLQSs245URa3x2HrISwp"

BASE_DRIVE = "/content/drive/MyDrive/YT_AUTOMATION"

PATHS = {
    "READY": os.path.join(BASE_DRIVE, "ready_to_upload"),
    "CHANNELS": os.path.join(BASE_DRIVE, "channels"),
    "CONFIG": os.path.join(BASE_DRIVE, "channels_config.json"),
    "PEXELS_RAW": os.path.join(BASE_DRIVE, "assets/pexels_raw"),
    "PEXELS_NORMAL": os.path.join(BASE_DRIVE, "assets/pexels_normalized"),
    "PROJECTS": os.path.join(BASE_DRIVE, "projects"),
    "OUTPUT": os.path.join(BASE_DRIVE, "output"),
    "ARCHIVE": os.path.join(BASE_DRIVE, "archive"),
    "COOKIES": os.path.join(BASE_DRIVE, "cookies.txt"),
    "LOGS": os.path.join(BASE_DRIVE, "activity_log.csv")
}

# Inisialisasi Folder
for p in PATHS.values():
    if not p.endswith(".json") and not p.endswith(".txt") and not p.endswith(".csv"):
        os.makedirs(p, exist_ok=True)

# Inisialisasi Log File
if not os.path.exists(PATHS["LOGS"]):
    pd.DataFrame(columns=["Waktu", "Kategori", "Pesan", "Status"]).to_csv(PATHS["LOGS"], index=False)

# Auto-Install Dependencies (Updated for google-genai)
try:
    import yt_dlp
    from google import genai
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "yt-dlp", "google-genai", "--upgrade", "--quiet"])
    import yt_dlp
    from google import genai

st.set_page_config(page_title="YT Command Center v8.10 (Quota Guard)", layout="wide", page_icon="üõ°Ô∏è")

# CSS Custom
st.markdown("""
<style>
    .stButton button {width: 100%; border-radius: 8px;}
    .card-container {background-color: #262730; padding: 15px; border-radius: 10px; margin-bottom: 10px; border: 1px solid #41444e;}
    div[data-testid="stExpander"] {border: none; box-shadow: none;}
    .quota-warning {color: #ff4b4b; font-weight: bold; border: 1px solid #ff4b4b; padding: 10px; border-radius: 5px;}
</style>
""", unsafe_allow_html=True)

# AI Client Setup (New SDK)
if GEMINI_API_KEY and "PASTE" not in GEMINI_API_KEY:
    try:
        ai_client = genai.Client(api_key=GEMINI_API_KEY)
        HAS_AI = True
    except:
        HAS_AI = False
else: HAS_AI = False

# =========================
# üìù SISTEM LOGGING
# =========================
def catat_log(kategori, pesan, status="INFO"):
    waktu = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    entry = pd.DataFrame([[waktu, kategori, pesan, status]], columns=["Waktu", "Kategori", "Pesan", "Status"])
    entry.to_csv(PATHS["LOGS"], mode='a', header=not os.path.exists(PATHS["LOGS"]), index=False)

def baca_log():
    if os.path.exists(PATHS["LOGS"]):
        return pd.read_csv(PATHS["LOGS"]).sort_values(by="Waktu", ascending=False)
    return pd.DataFrame()

# =========================
# üîß FUNGSI UTILITAS
# =========================

def sanitize_filename(name):
    return "".join([c if c.isalnum() else "_" for c in name]).strip("_")

def sanitize_tags(tags_list):
    if not tags_list or not isinstance(tags_list, list): return []
    valid_tags = []
    total_char = 0
    for tag in tags_list:
        clean_tag = str(tag).replace(",", "").strip()
        if total_char + len(clean_tag) + 1 > 450: break
        if len(clean_tag) > 1:
            valid_tags.append(clean_tag)
            total_char += len(clean_tag) + 1
    return valid_tags

def muat_konfigurasi():
    if os.path.exists(PATHS["CONFIG"]):
        try:
            with open(PATHS["CONFIG"], "r") as f: return json.load(f)
        except: return {}
    return {}

def simpan_konfigurasi(data):
    with open(PATHS["CONFIG"], "w") as f: json.dump(data, f, indent=4)
    catat_log("Config", "Konfigurasi diperbarui", "SUCCESS")

def dapatkan_durasi(file):
    try:
        cmd = ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", file]
        hasil = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        return float(hasil.stdout)
    except: return 0.0

def dapatkan_resolusi(file_path):
    try:
        cmd = ["ffprobe", "-v", "error", "-select_streams", "v:0", "-show_entries", "stream=width,height", "-of", "csv=s=x:p=0", file_path]
        res = subprocess.check_output(cmd).decode("utf-8").strip().split("x")
        if len(res) == 2: return int(res[0]), int(res[1])
    except: pass
    return 0, 0

def dapatkan_layanan(folder_channel):
    tp = os.path.join(PATHS["CHANNELS"], folder_channel, "token.pickle")
    if not os.path.exists(tp): return None
    try:
        with open(tp, "rb") as t: creds = pickle.load(t)
        if creds.expired and creds.refresh_token:
            creds.refresh(Request())
            with open(tp, "wb") as t: pickle.dump(creds, t)
        return build("youtube", "v3", credentials=creds)
    except Exception as e:
        catat_log("Auth", f"Gagal login {folder_channel}: {e}", "ERROR")
        return None

def dapatkan_layanan_analitik(folder_channel):
    tp = os.path.join(PATHS["CHANNELS"], folder_channel, "token.pickle")
    if not os.path.exists(tp): return None
    try:
        with open(tp, "rb") as t: creds = pickle.load(t)
        return build('youtubeAnalytics', 'v2', credentials=creds)
    except: return None

def ambil_statistik_mendalam(folder_channel, channel_id):
    stats = {"views": 0, "watch_h": 0.0, "subs": 0, "likes": 0, "comments": 0, "total_videos": 0, "scheduled_count": 0, "next_schedule": "-", "api_status": "OK"}
    if not channel_id or "UCxxxx" in channel_id: return stats

    try:
        service_data = dapatkan_layanan(folder_channel)
        service_analytics = dapatkan_layanan_analitik(folder_channel)

        if service_data:
            try:
                ch_res = service_data.channels().list(part='statistics', id=channel_id).execute()
                if ch_res.get('items'): stats["total_videos"] = int(ch_res['items'][0]['statistics']['videoCount'])

                search_res = service_data.search().list(forMine=True, type='video', part='id', maxResults=50).execute()
                v_ids = [item['id']['videoId'] for item in search_res.get('items', [])]
                if v_ids:
                    v_details = service_data.videos().list(part='status', id=",".join(v_ids)).execute()
                    jadwal = [v['status'].get('publishAt') for v in v_details.get('items', []) if v['status'].get('publishAt')]
                    stats["scheduled_count"] = len(jadwal)
                    if jadwal: stats["next_schedule"] = min(jadwal).replace("T", " ").replace("Z", "")[:16]
            except HttpError as e:
                if "quotaExceeded" in str(e):
                    stats["api_status"] = "QUOTA_EXCEEDED"
                    return stats # Stop here, don't try analytics

        if service_analytics and stats["api_status"] == "OK":
            end = datetime.date.today().strftime('%Y-%m-%d')
            start = (datetime.date.today() - datetime.timedelta(days=30)).strftime('%Y-%m-%d')
            try:
                ana_res = service_analytics.reports().query(ids=f'channel=={channel_id}', startDate=start, endDate=end, metrics='views,estimatedMinutesWatched,subscribersGained,likes,comments').execute()
                if "rows" in ana_res and ana_res["rows"]:
                    row = ana_res["rows"][0]
                    stats.update({"views": int(row[0]), "watch_h": float(round(row[1]/60, 1)), "subs": int(row[2]), "likes": int(row[3]), "comments": int(row[4])})
            except Exception: pass

    except Exception as e:
        if "quotaExceeded" in str(e): stats["api_status"] = "QUOTA_EXCEEDED"
        pass
    return stats

def ai_generate_maksimal(judul, deskripsi_lama=""):
    if not HAS_AI: return None
    # Updated for google.genai SDK
    prompt = f"""
    Bertindaklah sebagai Pakar SEO YouTube.
    Judul Asli: "{judul}"
    Konteks: "{deskripsi_lama[:200]}"

    Tugas:
    1. Judul Clickbait tapi relevan (max 60 char).
    2. Deskripsi SEO (3 paragraf).
    3. 15 Tag relevan (koma separated).

    Output JSON:
    {{ "title": "...", "description": "...", "tags": [...] }}
    """
    try:
        res = ai_client.models.generate_content(
            model='gemini-1.5-flash',
            contents=prompt
        )
        clean = res.text.replace("```json", "").replace("```", "").strip()
        return json.loads(clean)
    except: return None

# =========================
# üè≠ MESIN PRODUKSI
# =========================

def ambil_media_pexels(kw, limit, status_container):
    if not PEXELS_API_KEY or "PASTE" in PEXELS_API_KEY: return

    safe_kw = sanitize_filename(kw)
    raw_dir = os.path.join(PATHS["PEXELS_RAW"], safe_kw)
    norm_dir = os.path.join(PATHS["PEXELS_NORMAL"], safe_kw)
    os.makedirs(raw_dir, exist_ok=True)
    os.makedirs(norm_dir, exist_ok=True)

    headers = {"Authorization": PEXELS_API_KEY}
    page, collected = 1, 0

    while collected < limit:
        url = f"https://api.pexels.com/videos/search?query={kw}&per_page=40&page={page}&orientation=landscape"
        try:
            r = requests.get(url, headers=headers); data = r.json()
            if 'videos' not in data or not data['videos']: break

            for vid in data["videos"]:
                best = next((f for f in vid["video_files"] if f["width"] >= 1920), vid["video_files"][0])

                fname_raw = os.path.join(raw_dir, f"{vid['id']}.mp4")
                fname_norm = os.path.join(norm_dir, f"{vid['id']}.mp4")

                if not os.path.exists(fname_raw) and not os.path.exists(fname_norm):
                    with requests.get(best["link"], stream=True) as s, open(fname_raw, 'wb') as f:
                        shutil.copyfileobj(s.raw, f)

                collected += 1
                if collected >= limit: break
            page += 1
        except: break
    return safe_kw

def normalisasi_video_wrapper(file_path_raw):
    file_path_norm = file_path_raw.replace("pexels_raw", "pexels_normalized")
    os.makedirs(os.path.dirname(file_path_norm), exist_ok=True)

    # 1. SMART SKIP: Jika sudah ada dan valid, lewati (Instant)
    if os.path.exists(file_path_norm) and os.path.getsize(file_path_norm) > 10000:
        return # Skip

    # 2. PROSES NORMALISASI (Jika belum ada)
    # Kita gunakan thread queue dari CPU untuk mempercepat proses (ultrafast)
    subprocess.run([
        "ffmpeg", "-y", "-loglevel", "error",
        "-i", file_path_raw,
        "-vf", "scale=1920:1080:force_original_aspect_ratio=increase,crop=1920:1080,setsar=1",
        "-r", "30",              # Kunci FPS ke 30 (Penting untuk concat)
        "-c:v", "libx264",       # Codec standar
        "-preset", "ultrafast",  # Mode Tercepat
        "-threads", "0",         # Pakai semua inti CPU
        "-an",                   # Hapus audio bawaan pexels (biar ringan)
        file_path_norm
    ])

def unduh_sumber_dan_rakit_paralel(data_paket):
    link = data_paket['link']
    mode = data_paket['mode']
    keyword = data_paket['keyword']

    try:
        # --- TAHAP 1: DOWNLOAD YOUTUBE ---
        with yt_dlp.YoutubeDL({'quiet': True, 'extract_flat': True, 'noplaylist': True}) as ydl:
            info_basic = ydl.extract_info(link, download=False)
            real_id = info_basic.get('id')

        proj_dir = os.path.join(PATHS["PROJECTS"], real_id)
        os.makedirs(proj_dir, exist_ok=True)

        if os.path.exists(os.path.join(PATHS["READY"], real_id, "video.mp4")):
            return f"‚úÖ SKIP (Sudah Ada): {real_id}"

        extractor_args = {'youtube': {'player_client': ['ios', 'android', 'web']}}
        base_opts = {
            'outtmpl': os.path.join(proj_dir, 'source.%(ext)s'),
            'quiet': True, 'nocheckcertificate': True, 'noplaylist': True,
            'socket_timeout': 30, 'extractor_args': extractor_args
        }
        if os.path.exists(PATHS["COOKIES"]): base_opts['cookiefile'] = PATHS["COOKIES"]

        info = None
        if mode == "Stok Pexels (Auto)":
            opts_audio = base_opts.copy()
            opts_audio['format'] = 'bestaudio/best'
            opts_audio['outtmpl'] = os.path.join(proj_dir, 'audio.%(ext)s')
            opts_audio['writethumbnail'] = True
            opts_audio['postprocessors'] = [{'key': 'FFmpegExtractAudio','preferredcodec': 'm4a'}, {'key': 'FFmpegThumbnailsConvertor','format': 'jpg'}]
            with yt_dlp.YoutubeDL(opts_audio) as ydl: info = ydl.extract_info(link, download=True)
            for f in os.listdir(proj_dir):
                if f.endswith(".jpg") and "audio" in f: os.rename(os.path.join(proj_dir, f), os.path.join(proj_dir, "source_thumbnail.jpg"))
        else:
            try:
                opts_1 = base_opts.copy(); opts_1['format'] = 'bestvideo[height>=1080]+bestaudio'; opts_1['merge_output_format'] = 'mp4'
                with yt_dlp.YoutubeDL(opts_1) as ydl: info = ydl.extract_info(link, download=True)
            except:
                try:
                    opts_2 = base_opts.copy(); opts_2['format'] = 'bestvideo[height>=720]+bestaudio'; opts_2['merge_output_format'] = 'mp4'
                    with yt_dlp.YoutubeDL(opts_2) as ydl: info = ydl.extract_info(link, download=True)
                except:
                    opts_3 = base_opts.copy(); opts_3['format'] = 'best'
                    if 'merge_output_format' in opts_3: del opts_3['merge_output_format']
                    with yt_dlp.YoutubeDL(opts_3) as ydl: info = ydl.extract_info(link, download=True)

        if info:
            meta = {"id": real_id, "title": info['title'], "description": info.get("description", ""), "tags": info.get("tags", [])}
            with open(os.path.join(proj_dir, "metadata.json"), "w") as f: json.dump(meta, f, indent=4)

        # --- TAHAP 2: RAKIT VIDEO (COPY PASTE MODE) ---
        ready_dir = os.path.join(PATHS["READY"], real_id)
        os.makedirs(ready_dir, exist_ok=True)
        out_path = os.path.join(ready_dir, "video.mp4")

        if mode == "Video Asli":
            src_files = [f for f in os.listdir(proj_dir) if f.startswith("source.") and f != "source.mp4"]
            src = os.path.join(proj_dir, "source.mp4")
            final_src = src if os.path.exists(src) else (os.path.join(proj_dir, src_files[0]) if src_files else None)

            if final_src:
                w, h = dapatkan_resolusi(final_src)
                if h > 0 and h < 1080:
                    subprocess.run(["ffmpeg", "-y", "-loglevel", "error", "-i", final_src, "-vf", "scale=1920:1080:force_original_aspect_ratio=increase,crop=1920:1080,setsar=1", "-c:v", "libx264", "-preset", "ultrafast", "-c:a", "aac", out_path])
                else:
                    if final_src.endswith(".mp4") and h >= 1080: shutil.copy2(final_src, out_path)
                    else: subprocess.run(["ffmpeg", "-y", "-loglevel", "error", "-i", final_src, "-c:v", "libx264", "-preset", "ultrafast", "-c:a", "aac", out_path])
        else:
            # Pexels Mode (Copy Paste Murni)
            audio = next((os.path.join(proj_dir, f) for f in os.listdir(proj_dir) if f.endswith(('.m4a', '.webm', '.mp3'))), None)
            if audio:
                durasi = dapatkan_durasi(audio)
                klip = []
                if keyword:
                    safe_kw = sanitize_filename(keyword)
                    target_folder = os.path.join(PATHS["PEXELS_NORMAL"], safe_kw)
                    if os.path.exists(target_folder): klip = [os.path.join(target_folder, f) for f in os.listdir(target_folder) if f.endswith(".mp4")]

                if not klip:
                    for root, dirs, files in os.walk(PATHS["PEXELS_NORMAL"]):
                        for f in files:
                            if f.endswith(".mp4"): klip.append(os.path.join(root, f))

                if klip:
                    random.shuffle(klip)
                    list_txt = os.path.join(proj_dir, "concat.txt")
                    curr = 0
                    with open(list_txt, "w") as f:
                        while curr < durasi:
                            for c in klip:
                                d = dapatkan_durasi(c)
                                if d < 1: continue
                                f.write(f"file '{c}'\n"); curr += d
                                if curr >= durasi: break
                    # INI ADALAH BAGIAN COPY PASTE (SANGAT CEPAT)
                    subprocess.run(["ffmpeg", "-y", "-loglevel", "error", "-f", "concat", "-safe", "0", "-i", list_txt, "-i", audio, "-c:v", "copy", "-c:a", "copy", "-shortest", out_path])

        if os.path.exists(os.path.join(proj_dir, "metadata.json")): shutil.copy2(os.path.join(proj_dir, "metadata.json"), os.path.join(ready_dir, "metadata.json"))
        src_thumb = os.path.join(proj_dir, "source_thumbnail.jpg")
        final_thumb = os.path.join(ready_dir, "thumbnail.jpg")
        if os.path.exists(src_thumb): shutil.copy2(src_thumb, final_thumb)
        else:
            thumb = next((os.path.join(proj_dir, f) for f in os.listdir(proj_dir) if f.endswith((".jpg", ".webp")) and "source" in f), None)
            if thumb: shutil.copy2(thumb, final_thumb)

        return f"‚úÖ SUKSES: {real_id}"

    except Exception as e:
        return f"‚ùå GAGAL {link}: {str(e)}"

# =========================
# üì∫ TATA LETAK APLIKASI
# =========================

t1, t2, t3, t4, t5, t6 = st.tabs(["üìä Dasbor", "üè≠ Studio", "üì¶ Manajer", "üì§ Injeksi", "‚öôÔ∏è Config", "üìú Log"])

with t1:
    st.header("üì° Dasbor Armada")
    conf = muat_konfigurasi()
    if conf:
        data_armada = []
        for nama, v in conf.items():
            with st.spinner(f"Cek {nama}..."):
                stat = ambil_statistik_mendalam(v.get('folder_name'), v.get('channel_id'))

                # STATUS CHECK
                status_icon = "üü¢" if v.get("active") else "üî¥"
                if stat.get("api_status") == "QUOTA_EXCEEDED":
                    status_icon = "‚ö†Ô∏è QUOTA HABIS"
                    st.toast(f"Kuota Youtube API habis untuk {nama}. Tunggu reset besok.", icon="‚ö†Ô∏è")

                data_armada.append({
                    "Status": status_icon,
                    "Nama Channel": nama,
                    "Total Post": int(stat["total_videos"]),
                    "Views (30h)": f"{stat['views']:,}",
                    "Jam Tayang": float(stat["watch_h"]),
                    "Subs Baru": int(stat["subs"]),
                    "Antrean": int(stat["scheduled_count"]),
                    "Jadwal Berikutnya": str(stat["next_schedule"])
                })

        df_armada = pd.DataFrame(data_armada)
        if not df_armada.empty:
            df_armada["Total Post"] = df_armada["Total Post"].astype(int)
            st.dataframe(df_armada, width="stretch", hide_index=True)

            if any(d["Status"] == "‚ö†Ô∏è QUOTA HABIS" for d in data_armada):
                st.markdown('<div class="quota-warning">‚ö†Ô∏è PERINGATAN: Kuota Harian YouTube API Habis! Upload & Statistik dimatikan sementara. Fitur produksi video masih bisa digunakan.</div>', unsafe_allow_html=True)

    else: st.info("Belum ada channel.")

with t2:
    st.header("üè≠ Studio Produksi (Turbo Parallel)")
    c1, c2 = st.columns(2)
    with c1:
        mode_v = st.radio("Jalur Produksi:", ["Stok Pexels (Auto)", "Video Asli (Reupload)"])
        input_link = st.text_area("Link YouTube (Pisahkan baris):")
    with c2:
        col_kw, col_limit = st.columns([3, 1])
        with col_kw:
            kata_kunci = st.text_input("Kata Kunci Visual (Pexels):", "nature cinematic")
        with col_limit:
            limit_stok = st.number_input("Jml Stok:", min_value=5, max_value=100, value=30, step=5)

        st.write("---")
        workers_count = st.slider("Jumlah Thread (Kekuatan Paralel):", 1, 4, 2, help="Semakin tinggi makin cepat, tapi RAM makin berat.")

        if st.button("üöÄ MULAI PRODUKSI (TURBO)", type="primary"):
            daftar_link = [l.strip() for l in input_link.split("\n") if l.strip()]
            if not daftar_link: st.error("Link kosong!")
            else:
                with st.status("Sedang Memproses Secara Paralel...", expanded=True) as status:
                    # 1. PREPARE STOK (Serial - Sekali Saja per Keyword)
                    if mode_v == "Stok Pexels (Auto)":
                        status.write(f"üì• Cek stok untuk topik: '{kata_kunci}' (Target: {limit_stok} files)...")
                        folder_kw = ambil_media_pexels(kata_kunci, limit_stok, status)
                        raw_kw_path = os.path.join(PATHS["PEXELS_RAW"], folder_kw)
                        if os.path.exists(raw_kw_path):
                            file_raw_paths = [os.path.join(raw_kw_path, f) for f in os.listdir(raw_kw_path) if f.endswith(".mp4")]
                            if file_raw_paths:
                                status.write(f"üé® Normalisasi (Persiapan Bahan) {len(file_raw_paths)} video...")
                                with ThreadPoolExecutor(max_workers=os.cpu_count()) as ex: ex.map(normalisasi_video_wrapper, file_raw_paths)

                    # 2. EKSEKUSI UTAMA (PARALEL)
                    status.write(f"üöÄ Menjalankan {workers_count} mesin produksi sekaligus...")
                    paket_kerja = [{'link': l, 'mode': mode_v, 'keyword': kata_kunci} for l in daftar_link]

                    completed_count = 0
                    log_box = st.empty()

                    with ThreadPoolExecutor(max_workers=workers_count) as executor:
                        future_to_url = {executor.submit(unduh_sumber_dan_rakit_paralel, pkg): pkg for pkg in paket_kerja}

                        for future in as_completed(future_to_url):
                            res = future.result()
                            completed_count += 1
                            catat_log("TurboEngine", res, "INFO" if "SUKSES" in res else "ERROR")
                            log_box.write(f"‚úÖ Selesai ({completed_count}/{len(daftar_link)}): {res}")

                    status.update(label="‚úÖ Semua Tugas Selesai!", state="complete")
                st.success("Produksi Selesai! Cek Tab Manajer Massal."); time.sleep(3); st.rerun()

with t3:
    st.header("üì¶ Manajer Upload")
    item_siap = []
    if os.path.exists(PATHS["READY"]):
        for vid_id in sorted(os.listdir(PATHS["READY"])):
            path_meta = os.path.join(PATHS["READY"], vid_id, "metadata.json")
            if os.path.exists(path_meta):
                with open(path_meta, "r") as f: item_siap.append(json.load(f))

    st.write(f"Antrean: {len(item_siap)} Video")

    with st.expander("üöÄ Eksekusi Upload"):
        target_ch = st.selectbox("Channel Tujuan:", list(conf.keys()) if conf else [])
        if st.button("UPLOAD SEMUA"):
            # QUOTA CHECK
            try:
                layanan = dapatkan_layanan(conf[target_ch]["folder_name"])
                # Test call to check quota
                layanan.channels().list(part='id', mine=True).execute()
            except HttpError as e:
                if "quotaExceeded" in str(e):
                    st.error("‚ùå KUOTA HABIS! Tidak bisa upload sekarang. Tunggu reset besok.")
                    st.stop()
            except: pass

            with st.status("Uploading...") as status:
                for item in item_siap:
                    clean_tags = sanitize_tags(item['tags'])
                    body = {"snippet": {"title": item['title'], "description": item['description'], "tags": clean_tags, "categoryId": "22"}, "status": {"privacyStatus": "private"}}
                    try:
                        v_path = os.path.join(PATHS["READY"], item['id'], "video.mp4")
                        layanan.videos().insert(part="snippet,status", body=body, media_body=MediaFileUpload(v_path, resumable=True)).execute()
                        shutil.move(os.path.join(PATHS["READY"], item['id']), os.path.join(PATHS["ARCHIVE"], item['id']))
                    except Exception as e: st.error(f"Err {item['title']}: {e}")
                status.update(label="Selesai!", state="complete")

with t4:
    st.header("üì§ Injeksi Manual")
    fup = st.file_uploader("Video MP4")
    if fup and st.button("Simpan"):
        vid = f"man_{int(time.time())}"
        pth = os.path.join(PATHS["READY"], vid)
        os.makedirs(pth, exist_ok=True)
        with open(os.path.join(pth, "video.mp4"), "wb") as f: f.write(fup.getbuffer())
        with open(os.path.join(pth, "metadata.json"), "w") as f: json.dump({"id": vid, "title": fup.name, "description": "", "tags": []}, f)
        st.success("Tersimpan!"); st.rerun()

with t5:
    st.header("‚öôÔ∏è Config Editor")

    st.subheader("üßπ Pembersihan Sistem")
    st.caption("Gunakan ini jika video yang dihasilkan masih 720p (karena file lama).")
    if st.button("üóëÔ∏è Hapus Cache Normalisasi (Paksa 1080p Ulang)"):
        try:
            if os.path.exists(PATHS["PEXELS_NORMAL"]):
                shutil.rmtree(PATHS["PEXELS_NORMAL"])
                os.makedirs(PATHS["PEXELS_NORMAL"])
            st.success("Cache Normalisasi Dihapus! Silakan produksi ulang agar aset baru menjadi 1080p.")
        except Exception as e:
            st.error(f"Gagal hapus: {e}")

    st.markdown("---")

    # === FITUR BARU: CEK STATUS COOKIES ===
    st.subheader("üç™ Status Cookies (YouTube Download)")
    cookie_path = PATHS["COOKIES"]
    if os.path.exists(cookie_path):
        st.success(f"‚úÖ File cookies.txt DITEMUKAN di: {cookie_path}")

        # Cek isi file (validasi sederhana)
        try:
            with open(cookie_path, "r", encoding="utf-8", errors="ignore") as f:
                content = f.read(100) # Baca 100 karakter awal
                if "# Netscape HTTP Cookie File" in content or ".google.com" in content or ".youtube.com" in content:
                    st.info("‚ÑπÔ∏è Format file terlihat benar (Netscape/HTTP).")
                elif "{" in content and "}" in content:
                    st.error("‚ùå Peringatan: File terlihat seperti JSON! YT-DLP butuh format NETSCAPE.")
        except:
            st.warning("‚ö†Ô∏è Tidak bisa membaca isi file cookies.")

    else:
        st.error(f"‚ùå File cookies.txt TIDAK DITEMUKAN di: {cookie_path}")
        st.warning("üëâ Silakan upload file 'cookies.txt' (format Netscape) ke folder 'YT_AUTOMATION' di Google Drive Anda.")

    st.markdown("---")

    txt = st.text_area("JSON Config Channel", value=json.dumps(muat_konfigurasi(), indent=4))
    if st.button("Simpan Config"): simpan_konfigurasi(json.loads(txt)); st.rerun()

with t6:
    st.dataframe(baca_log(), width="stretch")
````
