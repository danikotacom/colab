````python
%%writefile app.py
import streamlit as st
import os, json, shutil, pickle, datetime, subprocess, random, requests, sys, time
import pandas as pd
import google.generativeai as genai
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from google.auth.transport.requests import Request
from PIL import Image
from concurrent.futures import ThreadPoolExecutor

# =========================
# ğŸ”‘ KONFIGURASI GLOBAL
# =========================
GEMINI_API_KEY = "PASTE_GEMINI_KEY_DISINI"
PEXELS_API_KEY = "PASTE_PEXELS_KEY_DISINI"

BASE_DRIVE = "/content/drive/MyDrive/YT_AUTOMATION"

PATHS = {
    "READY": os.path.join(BASE_DRIVE, "ready_to_upload"),
    "CHANNELS": os.path.join(BASE_DRIVE, "channels"),
    "CONFIG": os.path.join(BASE_DRIVE, "channels_config.json"),
    "PEXELS_RAW": os.path.join(BASE_DRIVE, "assets/pexels_raw"),
    "PEXELS_NORMAL": os.path.join(BASE_DRIVE, "assets/pexels_normalized"),
    "PROJECTS": os.path.join(BASE_DRIVE, "projects"),
    "OUTPUT": os.path.join(BASE_DRIVE, "output"),
    "ARCHIVE": os.path.join(BASE_DRIVE, "archive"),
    "COOKIES": os.path.join(BASE_DRIVE, "cookies.txt"),
    "LOGS": os.path.join(BASE_DRIVE, "activity_log.csv")
}

# Inisialisasi Folder
for p in PATHS.values():
    if not p.endswith(".json") and not p.endswith(".txt") and not p.endswith(".csv"):
        os.makedirs(p, exist_ok=True)

# Inisialisasi Log File jika belum ada
if not os.path.exists(PATHS["LOGS"]):
    pd.DataFrame(columns=["Waktu", "Kategori", "Pesan", "Status"]).to_csv(PATHS["LOGS"], index=False)

try:
    import yt_dlp
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "yt-dlp", "--upgrade", "--quiet"])
    import yt_dlp

st.set_page_config(page_title="YT Command Center v7.5", layout="wide", page_icon="ğŸ“¡")

# CSS Custom untuk Tampilan Kartu
st.markdown("""
<style>
    .stButton button {width: 100%; border-radius: 8px;}
    .card-container {
        background-color: #262730;
        padding: 15px;
        border-radius: 10px;
        margin-bottom: 10px;
        border: 1px solid #41444e;
    }
    div[data-testid="stExpander"] {border: none; box-shadow: none;}
</style>
""", unsafe_allow_html=True)

if GEMINI_API_KEY and "PASTE" not in GEMINI_API_KEY:
    genai.configure(api_key=GEMINI_API_KEY)
    HAS_AI = True
else: HAS_AI = False

# =========================
# ğŸ“ SISTEM LOGGING
# =========================
def catat_log(kategori, pesan, status="INFO"):
    waktu = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    entry = pd.DataFrame([[waktu, kategori, pesan, status]], columns=["Waktu", "Kategori", "Pesan", "Status"])
    # Append ke CSV tanpa load seluruh file (lebih cepat)
    entry.to_csv(PATHS["LOGS"], mode='a', header=not os.path.exists(PATHS["LOGS"]), index=False)

def baca_log():
    if os.path.exists(PATHS["LOGS"]):
        return pd.read_csv(PATHS["LOGS"]).sort_values(by="Waktu", ascending=False)
    return pd.DataFrame()

# =========================
# ğŸ”§ FUNGSI UTILITAS
# =========================

def sanitize_tags(tags_list):
    """Membersihkan tag agar tidak melebihi batas 500 karakter YouTube"""
    if not tags_list or not isinstance(tags_list, list): return []
    valid_tags = []
    total_char = 0
    for tag in tags_list:
        clean_tag = str(tag).replace(",", "").strip()
        if total_char + len(clean_tag) + 1 > 450:
            break
        if len(clean_tag) > 1:
            valid_tags.append(clean_tag)
            total_char += len(clean_tag) + 1
    return valid_tags

def muat_konfigurasi():
    if os.path.exists(PATHS["CONFIG"]):
        try:
            with open(PATHS["CONFIG"], "r") as f:
                return json.load(f)
        except:
            return {}
    return {}

def simpan_konfigurasi(data):
    with open(PATHS["CONFIG"], "w") as f:
        json.dump(data, f, indent=4)
    catat_log("Config", "Konfigurasi diperbarui", "SUCCESS")

def dapatkan_durasi(file):
    try:
        cmd = ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", file]
        hasil = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        return float(hasil.stdout)
    except: return 0.0

def dapatkan_layanan(folder_channel):
    tp = os.path.join(PATHS["CHANNELS"], folder_channel, "token.pickle")
    if not os.path.exists(tp): return None
    try:
        with open(tp, "rb") as t: creds = pickle.load(t)
        if creds.expired and creds.refresh_token:
            creds.refresh(Request())
            with open(tp, "wb") as t: pickle.dump(creds, t)
        return build("youtube", "v3", credentials=creds)
    except Exception as e:
        catat_log("Auth", f"Gagal login {folder_channel}: {e}", "ERROR")
        return None

def dapatkan_layanan_analitik(folder_channel):
    tp = os.path.join(PATHS["CHANNELS"], folder_channel, "token.pickle")
    if not os.path.exists(tp): return None
    try:
        with open(tp, "rb") as t: creds = pickle.load(t)
        return build('youtubeAnalytics', 'v2', credentials=creds)
    except: return None

def ambil_statistik_mendalam(folder_channel, channel_id):
    stats = {"views": 0, "watch_h": 0.0, "subs": 0, "likes": 0, "comments": 0, "total_videos": 0, "scheduled_count": 0, "next_schedule": "-"}
    if not channel_id or "UCxxxx" in channel_id: return stats

    try:
        service_data = dapatkan_layanan(folder_channel)
        service_analytics = dapatkan_layanan_analitik(folder_channel)

        if service_data:
            ch_res = service_data.channels().list(part='statistics', id=channel_id).execute()
            if ch_res.get('items'): stats["total_videos"] = int(ch_res['items'][0]['statistics']['videoCount'])

            search_res = service_data.search().list(forMine=True, type='video', part='id', maxResults=50).execute()
            v_ids = [item['id']['videoId'] for item in search_res.get('items', [])]
            if v_ids:
                v_details = service_data.videos().list(part='status', id=",".join(v_ids)).execute()
                jadwal = [v['status'].get('publishAt') for v in v_details.get('items', []) if v['status'].get('publishAt')]
                stats["scheduled_count"] = len(jadwal)
                if jadwal: stats["next_schedule"] = min(jadwal).replace("T", " ").replace("Z", "")[:16]

        if service_analytics:
            end = datetime.date.today().strftime('%Y-%m-%d')
            start = (datetime.date.today() - datetime.timedelta(days=30)).strftime('%Y-%m-%d')
            ana_res = service_analytics.reports().query(ids=f'channel=={channel_id}', startDate=start, endDate=end, metrics='views,estimatedMinutesWatched,subscribersGained,likes,comments').execute()
            if "rows" in ana_res and ana_res["rows"]:
                row = ana_res["rows"][0]
                stats.update({"views": int(row[0]), "watch_h": float(round(row[1]/60, 1)), "subs": int(row[2]), "likes": int(row[3]), "comments": int(row[4])})
    except: pass
    return stats

def ai_generate_maksimal(judul, deskripsi_lama=""):
    if not HAS_AI: return None
    model = genai.GenerativeModel('gemini-1.5-flash')
    prompt = f"""
    Bertindaklah sebagai Pakar SEO YouTube & Copywriter Handal.
    Judul Asli Video: "{judul}"
    Konteks Awal: "{deskripsi_lama[:200]}"

    Tugas:
    1. Buat Judul yang Sangat Menarik (Clickbait tapi relevan, max 60 char).
    2. Buat Deskripsi SEO-friendly (Minimal 3 paragraf, sertakan kata kunci).
    3. Buat 15 Tag yang relevan (dipisah koma).

    Output WAJIB JSON Valid:
    {{
        "title": "Judul Baru",
        "description": "Deskripsi Baru...",
        "tags": ["tag1", "tag2"]
    }}
    """
    try:
        res = model.generate_content(prompt)
        clean = res.text.replace("```json", "").replace("```", "").strip()
        return json.loads(clean)
    except Exception as e:
        catat_log("AI", f"Gagal generate AI: {e}", "ERROR")
        return None

# =========================
# ğŸ­ MESIN PRODUKSI
# =========================

def unduh_sumber(link, status_container, mode):
    proj_id = "".join(filter(str.isalnum, link))[-12:]
    proj_dir = os.path.join(PATHS["PROJECTS"], proj_id)
    os.makedirs(proj_dir, exist_ok=True)

    opts = {
        'format': 'bestvideo[height<=1080]+bestaudio/best' if mode == "Video Asli" else 'bestaudio/best',
        'outtmpl': os.path.join(proj_dir, 'source.%(ext)s'),
        'merge_output_format': 'mp4', 'quiet': True, 'nocheckcertificate': True,
        'user_agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'
    }
    if os.path.exists(PATHS["COOKIES"]): opts['cookiefile'] = PATHS["COOKIES"]

    try:
        with yt_dlp.YoutubeDL(opts) as ydl:
            info = ydl.extract_info(link, download=True)
            meta = {"id": proj_id, "title": info['title'], "description": info.get("description", ""), "tags": info.get("tags", [])}
            with open(os.path.join(proj_dir, "metadata.json"), "w") as f: json.dump(meta, f, indent=4)
            catat_log("Creator", f"Berhasil unduh: {info['title']}", "SUCCESS")
            return proj_id
    except Exception as e:
        status_container.error(f"Gagal Unduh: {e}")
        catat_log("Creator", f"Gagal unduh {link}: {e}", "ERROR")
        return None

def bangun_video(vid_id, mode, status_container):
    proj_dir, ready_dir = os.path.join(PATHS["PROJECTS"], vid_id), os.path.join(PATHS["READY"], vid_id)
    os.makedirs(ready_dir, exist_ok=True)
    out_path = os.path.join(ready_dir, "video.mp4")

    try:
        if mode == "Video Asli":
            src = os.path.join(proj_dir, "source.mp4")
            if os.path.exists(src): shutil.copy2(src, out_path)
        else:
            audio = next((os.path.join(proj_dir, f) for f in os.listdir(proj_dir) if f.endswith(('.m4a', '.webm', '.mp3'))), None)
            if not audio: return
            durasi = dapatkan_durasi(audio)
            klip = [os.path.join(PATHS["PEXELS_NORMAL"], f) for f in os.listdir(PATHS["PEXELS_NORMAL"]) if f.endswith(".mp4")]
            if not klip: return
            random.shuffle(klip)
            list_txt = os.path.join(proj_dir, "concat.txt")
            curr = 0
            with open(list_txt, "w") as f:
                while curr < durasi:
                    for c in klip:
                        d = dapatkan_durasi(c)
                        if d < 1: continue
                        f.write(f"file '{c}'\n"); curr += d
                        if curr >= durasi: break
            subprocess.run(["ffmpeg", "-y", "-loglevel", "error", "-f", "concat", "-safe", "0", "-i", list_txt, "-i", audio, "-c:v", "copy", "-c:a", "copy", "-shortest", out_path])

        if os.path.exists(os.path.join(proj_dir, "metadata.json")):
            shutil.copy2(os.path.join(proj_dir, "metadata.json"), os.path.join(ready_dir, "metadata.json"))

        thumb_src = os.path.join(proj_dir, "source.jpg")
        if not os.path.exists(thumb_src):
             for f in os.listdir(proj_dir):
                 if f.endswith(".jpg") or f.endswith(".webp"):
                     thumb_src = os.path.join(proj_dir, f)
                     break

        if os.path.exists(thumb_src):
            shutil.copy2(thumb_src, os.path.join(ready_dir, "thumbnail.jpg"))

        catat_log("Creator", f"Video {vid_id} siap diproses", "SUCCESS")
    except Exception as e:
        catat_log("Creator", f"Gagal Build {vid_id}: {e}", "ERROR")

def ambil_media_pexels(keyword, limit, status_container):
    if not PEXELS_API_KEY or "PASTE" in PEXELS_API_KEY: return
    headers = {"Authorization": PEXELS_API_KEY}
    page, collected = 1, 0
    while collected < limit:
        url = f"https://api.pexels.com/videos/search?query={keyword}&per_page=40&page={page}&orientation=landscape"
        try:
            r = requests.get(url, headers=headers); data = r.json()
            for vid in data["videos"]:
                best = next((f for f in vid["video_files"] if f["width"] >= 1920), vid["video_files"][0])
                fname = os.path.join(PATHS["PEXELS_RAW"], f"{vid['id']}.mp4")
                if not os.path.exists(fname):
                    with requests.get(best["link"], stream=True) as s, open(fname, 'wb') as f: shutil.copyfileobj(s.raw, f)
                collected += 1
                if collected >= limit: break
            page += 1
        except: break

def normalisasi_video(file_name):
    raw, norm = os.path.join(PATHS["PEXELS_RAW"], file_name), os.path.join(PATHS["PEXELS_NORMAL"], file_name)
    if os.path.exists(norm) and os.path.getsize(norm) > 10000: return
    subprocess.run(["ffmpeg", "-y", "-loglevel", "error", "-i", raw, "-vf", "scale=1920:1080:force_original_aspect_ratio=increase,crop=1920:1080,setsar=1", "-r", "30", "-c:v", "libx264", "-preset", "ultrafast", "-an", norm])

# =========================
# ğŸ“º TATA LETAK APLIKASI
# =========================

t1, t2, t3, t4, t5, t6 = st.tabs(["ğŸ“Š Dasbor Armada", "ğŸ­ Studio Kreator", "ğŸ“¦ Manajer Massal", "ğŸ“¤ Injeksi Manual", "âš™ï¸ Konfigurasi", "ğŸ“œ Riwayat"])

# --- TAB 1: DASBOR ---
with t1:
    st.header("ğŸ“¡ Dasbor Armada (Real-Time)")
    conf = muat_konfigurasi()
    if conf:
        data_armada = []
        for nama, v in conf.items():
            with st.spinner(f"Sinkronisasi {nama}..."):
                stat = ambil_statistik_mendalam(v.get('folder_name'), v.get('channel_id'))
                data_armada.append({
                    "Status": "ğŸŸ¢" if v.get("active") else "ğŸ”´",
                    "Nama Channel": nama,
                    "Total Post": int(stat["total_videos"]),
                    "Views (30h)": f"{stat['views']:,}",
                    "Jam Tayang": float(stat["watch_h"]),
                    "Subs Baru": int(stat["subs"]),
                    "Antrean": int(stat["scheduled_count"]),
                    "Jadwal Berikutnya": str(stat["next_schedule"])
                })

        df_armada = pd.DataFrame(data_armada)
        if not df_armada.empty:
            df_armada["Total Post"] = df_armada["Total Post"].astype(int)
            df_armada["Subs Baru"] = df_armada["Subs Baru"].astype(int)
            st.dataframe(df_armada, width="stretch", hide_index=True)
    else: st.info("âš ï¸ Konfigurasi Kosong.")

# --- TAB 2: PABRIK VIDEO ---
with t2:
    st.header("ğŸ­ Studio Produksi")
    c1, c2 = st.columns(2)
    with c1:
        mode_v = st.radio("Jalur Produksi:", ["Stok Pexels (Aman)", "Video Asli (Langsung)"])
        input_link = st.text_area("Tautan YouTube (Satu per baris):")
    with c2:
        kata_kunci = st.text_input("Kata Kunci Pexels:", "nature cinematic")
        if st.button("ğŸš€ MULAI PRODUKSI", type="primary"):
            daftar_link = [l.strip() for l in input_link.split("\n") if l.strip()]
            if not daftar_link: st.error("Masukkan link YouTube dulu!")
            else:
                with st.status("Sedang Memproses Produksi...", expanded=True) as status:
                    if mode_v == "Stok Pexels (Aman)":
                        status.write("ğŸ“¥ Mengambil Stok Pexels...")
                        ambil_media_pexels(kata_kunci, 30, status)
                        file_raw = [f for f in os.listdir(PATHS["PEXELS_RAW"]) if f.endswith(".mp4")]
                        status.write(f"ğŸ¨ Menormalisasi {len(file_raw)} video...")
                        with ThreadPoolExecutor(max_workers=4) as ex: ex.map(normalisasi_video, file_raw)

                    for i, l in enumerate(daftar_link):
                        status.write(f"â³ ({i+1}/{len(daftar_link)}) Memproses: {l}")
                        vid_id = unduh_sumber(l, status, "Video Asli" if "Asli" in mode_v else "Stok Pexels")
                        if vid_id: bangun_video(vid_id, "Video Asli" if "Asli" in mode_v else "Stok Pexels", status)
                    status.update(label="âœ… Produksi Batch Selesai!", state="complete", expanded=False)

                st.success("Semua video siap di Tab Manajer Massal!")
                time.sleep(2)
                st.rerun()

# --- TAB 3: MANAJER MASSAL ---
with t3:
    st.header("ğŸ“¦ Manajer Upload Massal")

    item_siap = []
    if os.path.exists(PATHS["READY"]):
        for vid_id in sorted(os.listdir(PATHS["READY"])):
            path_folder = os.path.join(PATHS["READY"], vid_id)
            path_meta = os.path.join(path_folder, "metadata.json")
            path_thumb = os.path.join(path_folder, "thumbnail.jpg")
            path_video = os.path.join(path_folder, "video.mp4")

            if os.path.exists(path_meta):
                with open(path_meta, "r") as f: d = json.load(f)
                item_siap.append({
                    "id": vid_id,
                    "title": d.get("title", "No Title"),
                    "desc": d.get("description", ""),
                    "tags": d.get("tags", []),
                    "thumb": path_thumb if os.path.exists(path_thumb) else None,
                    "video": path_video
                })

    with st.expander("ğŸ› ï¸ Pengaturan Upload Global", expanded=True):
        col_set1, col_set2, col_set3 = st.columns(3)
        with col_set1:
            target_ch = st.selectbox("Target Channel:", list(conf.keys()) if conf else [])
        with col_set2:
            interval = st.selectbox("Jadwal Otomatis:", ["Manual (Private)", "8 Jam (3x Sehari)", "24 Jam (1x Sehari)"])
        with col_set3:
            tgl_mulai = st.date_input("Mulai Tanggal:", datetime.date.today())
            jam_mulai = st.time_input("Mulai Jam:", datetime.time(17, 0))

    col_act1, col_act2 = st.columns([1, 4])
    with col_act1:
        if st.button("âœ¨ AI GENERATE ALL"):
            with st.status("Generating Metadata AI...", expanded=True):
                for item in item_siap:
                    st.write(f"Thinking for: {item['title']}")
                    ai_res = ai_generate_maksimal(item['title'], item['desc'])
                    if ai_res:
                        path_meta = os.path.join(PATHS["READY"], item['id'], "metadata.json")
                        with open(path_meta, "w") as f: json.dump(ai_res, f, indent=4)
            st.success("AI Selesai!"); st.rerun()

    with col_act2:
        if st.button("ğŸš€ UPLOAD SEMUA LIST DI BAWAH", type="primary"):
            if not item_siap: st.warning("List Kosong!")
            else:
                layanan = dapatkan_layanan(conf[target_ch]["folder_name"])
                waktu_skrg = datetime.datetime.combine(tgl_mulai, jam_mulai)
                gap = 8 if "8 Jam" in interval else 24

                with st.status("Mengunggah Video...", expanded=True) as status:
                    for item in item_siap:
                        status.write(f"ğŸš€ Uploading: {item['title']}...")
                        folder = os.path.join(PATHS["READY"], item['id'])
                        pub_at = waktu_skrg.isoformat() + "Z" if "Manual" not in interval else None
                        clean_tags = sanitize_tags(item['tags'])

                        body = {
                            "snippet": {"title": item['title'], "description": item['desc'], "tags": clean_tags, "categoryId": "22"},
                            "status": {"privacyStatus": "private", "selfDeclaredMadeForKids": False}
                        }
                        if pub_at: body["status"]["publishAt"] = pub_at

                        try:
                            req = layanan.videos().insert(part="snippet,status", body=body, media_body=MediaFileUpload(item['video'], resumable=True))
                            res = None
                            while res is None: _, res = req.next_chunk()

                            if item['thumb']:
                                layanan.thumbnails().set(videoId=res["id"], media_body=MediaFileUpload(item['thumb'])).execute()

                            shutil.move(folder, os.path.join(PATHS["ARCHIVE"], item['id']))
                            catat_log("Upload", f"Sukses upload {item['title']} ke {target_ch}", "SUCCESS")
                            waktu_skrg += datetime.timedelta(hours=gap)
                        except Exception as e:
                            catat_log("Upload", f"Gagal upload {item['title']}: {e}", "ERROR")
                            st.error(f"Gagal: {e}")
                    status.update(label="Selesai!", state="complete")
                st.balloons(); time.sleep(2); st.rerun()

    st.markdown("---")

    if not item_siap:
        st.info("ğŸ“­ Tidak ada video di antrian.")
    else:
        st.write(f"**Menampilkan {len(item_siap)} Video Siap Upload:**")
        for item in item_siap:
            with st.container():
                c_thumb, c_detail, c_action = st.columns([2, 4, 2])
                with c_thumb:
                    if item['thumb']: st.image(item['thumb'], use_container_width=True)
                    else: st.write("No Image")
                    with st.expander("â–¶ï¸ Putar Video"): st.video(item['video'])

                with c_detail:
                    st.subheader(item['title'])
                    st.caption(f"ID: {item['id']}")
                    with st.expander("ğŸ“ Edit Metadata"):
                        new_title = st.text_input("Judul", value=item['title'], key=f"t_{item['id']}")
                        new_desc = st.text_area("Deskripsi", value=item['desc'], height=150, key=f"d_{item['id']}")
                        if st.button("Simpan Perubahan", key=f"s_{item['id']}"):
                            meta = {"title": new_title, "description": new_desc, "tags": item['tags']}
                            with open(os.path.join(PATHS["READY"], item['id'], "metadata.json"), "w") as f:
                                json.dump(meta, f)
                            st.success("Disimpan!"); st.rerun()

                with c_action:
                    st.write("### Aksi")
                    if st.button("ğŸ—‘ï¸ HAPUS", key=f"del_{item['id']}", type="secondary"):
                        shutil.rmtree(os.path.join(PATHS["READY"], item['id']))
                        catat_log("Hapus", f"Video {item['id']} dihapus manual", "INFO")
                        st.warning("Dihapus."); st.rerun()
            st.markdown("---")

# --- TAB 4: INJEKSI MANUAL ---
with t4:
    st.header("ğŸ“¤ Injeksi Video Manual")
    file_up = st.file_uploader("Unggah Video MP4:")
    if st.button("Suntik ke Antrean"):
        if file_up:
            v_id = f"manual_{int(time.time())}"
            dir_tujuan = os.path.join(PATHS["READY"], v_id)
            os.makedirs(dir_tujuan, exist_ok=True)
            with open(os.path.join(dir_tujuan, "video.mp4"), "wb") as f: f.write(file_up.getbuffer())
            with open(os.path.join(dir_tujuan, "metadata.json"), "w") as f: json.dump({"id": v_id, "title": file_up.name, "description": "", "tags": []}, f)
            catat_log("Inject", f"Injeksi manual: {file_up.name}", "SUCCESS")
            st.success("Video berhasil masuk antrean!"); st.rerun()

# --- TAB 5: EDITOR KONFIGURASI ---
with t5:
    st.header("âš™ï¸ Editor Konfigurasi")
    try:
        current_conf = muat_konfigurasi()
        text_conf = st.text_area("JSON Config:", value=json.dumps(current_conf, indent=4), height=400)
        if st.button("ğŸ’¾ SIMPAN KONFIGURASI"):
            try:
                baru = json.loads(text_conf)
                simpan_konfigurasi(baru)
                st.success("Tersimpan!"); st.rerun()
            except: st.error("JSON Tidak Valid!")
    except: st.error("File Config Rusak/Hilang.")

# --- TAB 6: RIWAYAT ---
with t6:
    st.header("ğŸ“œ Riwayat Aktivitas")
    df_log = baca_log()
    if not df_log.empty:
        st.dataframe(df_log, width="stretch", hide_index=True)
        if st.button("Bersihkan Log"):
            os.remove(PATHS["LOGS"])
            st.success("Log bersih."); st.rerun()
    else: st.info("Belum ada aktivitas.")
````
